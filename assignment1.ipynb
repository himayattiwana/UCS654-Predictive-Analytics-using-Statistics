```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", encoding="latin1", low_memory=False)

df['no2'] = pd.to_numeric(df['no2'], errors='coerce')

df = df.dropna(subset=['no2'])

X = df['no2'].values    

r = 102313049

a_r = 0.05 * (r % 7)
b_r = 0.3 * ((r % 5) + 1)

z = X + a_r * np.sin(b_r * X)

mu = np.mean(z)
sigma = np.std(z)

lambda_ = 1 / (2 * sigma**2)
c = 1 / np.sqrt(2 * np.pi * sigma**2)

print("Learned parameters:")
print("mu =", mu)
print("lambda =", lambda_)
print("c =", c)
```
